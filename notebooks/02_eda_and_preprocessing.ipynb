# Exploratory Data Analysis - Exam Anxiety Dataset

This notebook performs EDA on the generated synthetic dataset of student exam reflections to understand label distributions and text characteristics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")

# Load dataset
df = pd.read_csv('../data/exam_anxiety_dataset.csv')
df.head()
```

## 1. Class Distribution
Checking if our classes (Low, Moderate, High anxiety) are balanced.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='label_text', palette='viridis', order=['Low', 'Moderate', 'High'])
plt.title('Distribution of Anxiety Levels')
plt.xlabel('Anxiety Level')
plt.ylabel('Count')
plt.show()

print(df['label_text'].value_counts())
```

## 2. Text Length Analysis
Analyzing the number of words in each category.

```python
df['word_count'] = df['text'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='label_text', y='word_count', palette='viridis', order=['Low', 'Moderate', 'High'])
plt.title('Word Count Distribution by Anxiety Level')
plt.xlabel('Anxiety Level')
plt.ylabel('Word Count')
plt.show()
```
